```python
#Load the required libraries
import pandas as pd
import numpy as np
import seaborn as sns

#Load the data
df = pd.read_csv('master.csv')

#View the data
canadaData = df[df['country'] == 'Canada']

print(canadaData)
```

```python
canadaData.to_csv('newfiledata.csv', sep='\t')
```

```python
Result = canadaData.groupby(['age','sex','generation'])['suicides_no'].sum()
print(Result)
Result.plot(kind='barh')
```

```python
import matplotlib.pyplot as plt
Result = canadaData.groupby(by="age")["suicides_no"].sum()
print(Result)
Result.plot(kind='bar')
```

```python
Result = canadaData.groupby(['generation'])['suicides_no'].sum()
print(Result)
Result.plot(kind='bar')
```
